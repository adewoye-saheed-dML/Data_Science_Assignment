## Data Cleaning and Preprocessing

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the dataset
df = pd.read_csv('C:/Users/WOYES/Downloads/study/study_performance.csv')
```

```python
# Check the top 5 rows of the dataset
df.head()
```

```python
# Check the number of observations and columns of the dataset
df.shape
```

```python
# Check for general information about the columns in the dataset
df.info()
```

```python
# Check for missing values
df.isna().sum()
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Check the number of unique values of each column
df.nunique()
```

```python
# Check the columns of the dataset
df.columns
```

```python
# Rename each column 
df=df.rename(columns={'gender':'Gender','race_ethnicity':'Race','parental_level_of_education':'Parent_education','lunch':'Lunch','test_preparation_course':'Test_prep','math_score':'Math_score','reading_score':'Reading_score','writing_score':'Writing_score'})
df.head()
```

```python
# Find the average score for the three tests
df['Total']=df['Math_score']+df['Reading_score']+df['Writing_score']
df['Average_score']=(df['Total']/3).round(2)
df.head()
```

```python
# Check the new shape of the dataset
df.shape
```

```python
# Create passed students dataframe (assuming the pass score is 60)
pass_students=df[df['Average_score']>=60]
pass_students.shape
```

## Data Visualization

```python
# Visualizing number of pass student by gender

cut_off=pass_students.groupby('Gender')['Average_score'].count()

fig, ax=plt.subplots()
ax.bar(cut_off.index, cut_off.values)

plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.title('Number of Students Who scored above 60 by Gender')
plt.show()
```

```python
# Visualize the passed Students Gender Distributon

gender_distr=pass_students['Gender'].value_counts()

fig, ax=plt.subplots()
ax.pie(gender_distr, labels=gender_distr.index, autopct='%1.1f%%')

plt.title('Passed Students Gender Distributon')

plt.axis('equal')
plt.show()
```

```python
# Visualize the passed students parent education distribution by gender

parent_education_levels = pass_students['Parent_education'].unique()

# Plot histogram for each gender
fig, ax = plt.subplots(2,1, figsize=(8, 6), sharex=True)

for i, gender in enumerate(['male', 'female']):
    df_gender = pass_students[pass_students['Gender'] == gender]
    ax[i].hist(df_gender['Parent_education'], bins=len(parent_education_levels), edgecolor='black', alpha=0.7, label=gender, stacked=True)

# Add labels and title
plt.suptitle('Passed_Students Parent Education Distribution by Gender')
plt.xlabel('Parent Education')
plt.ylabel('Counts')

# Set xticks and labels
plt.xticks(rotation=90)

# Add legend
plt.legend()
```

```python
# Visualize the passed_students gender's average score by parent education

pass_male = pass_students[pass_students['Gender']=='male']
pass_female = pass_students[pass_students['Gender']=='female']

# Creating the two boxplots
fig, ax = plt.subplots(2, 1, figsize=(10, 10))  

# Box plot for males
ax[0].boxplot([pass_male[pass_male['Parent_education'] == edu]['Average_score'] for edu in ["some high school", "high school", "some college", "associate's degree", "bachelor's degree", "master's degree"]], 
            labels=["some high school", "high school", "some college", "associate's degree", "bachelor's degree", "master's degree"])
ax[0].set_title('Distribution of Average Scores for Males with Pass Score by Parent Education')
ax[0].set_ylabel('Average Score')
ax[0].tick_params(axis='x', rotation=90) 
ax[0].grid(True)

# Box plot for females
ax[1].boxplot([pass_female[pass_female['Parent_education'] == edu2]['Average_score'] for edu2 in ["some high school", "high school", "some college", "associate's degree", "bachelor's degree", "master's degree"]], 
            labels=["some high school", "high school", "some college", "associate's degree", "bachelor's degree", "master's degree"])
ax[1].set_title('Distribution of Average Scores for Females with Pass Score by Parent Education')
ax[1].set_xlabel('Parent Education')
ax[1].set_ylabel('Average Score')
ax[1].tick_params(axis='x', rotation=90)  
ax[1].grid(True)

plt.tight_layout()  
plt.show()
```

```python
# Visualizing number of passed students that completed his/her test materials

male = pass_students[(pass_students['Gender']=='male')& (pass_students['Test_prep']=='completed')]
female = pass_students[(pass_students['Gender']=='female')& (pass_students['Test_prep']=='completed')]


labels = ['Male', 'Female']
completed_test_prep = [len(male), len(female)]

# Plotting
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(labels, completed_test_prep, color=['blue', 'orange'])

# Adding labels and title
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.title('Number of Students Who Completed Test Preparation by Gender')

# Display the plot
plt.show()
```

```python
# Number of passed male and female students who completed their test materials before test
male = len(pass_students[(pass_students['Gender']=='male')& (pass_students['Test_prep']=='completed')])
female = len(pass_students[(pass_students['Gender']=='female')& (pass_students['Test_prep']=='completed')])

print("Number of passed male student who completed their test materials:", male)
print("Number of passed female student who completed their test materials:",female)
```

```python
# Visualizing number of passed_students by gender who took lunch before test

genders = ['Male', 'Female']

# Number of male and female students with free/reduced lunch
male_free_counts = len(pass_students[(pass_students['Gender']=='male') & (pass_students['Lunch'] == 'free/reduced')])
female_free_counts = len(pass_students[(pass_students['Gender']=='female') & (pass_students['Lunch'] == 'free/reduced')])

# Number of male and female students with standard lunch
male_standard_counts = len(pass_students[(pass_students['Gender']=='male') & (pass_students['Lunch'] == 'standard')])
female_standard_counts = len(pass_students[(pass_students['Gender']=='female') & (pass_students['Lunch'] == 'standard')])

# Stacked bars for the passed students
plt.figure(figsize=(10, 6))

plt.bar(genders, [male_free_counts, female_free_counts], label='free/reduced', color='blue')
plt.bar(genders, [male_standard_counts, female_standard_counts], bottom=[male_free_counts, female_free_counts], label='standard', color='orange')

# Adding labels and title
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.title('Distribution of Lunch Types by Gender')
plt.legend()

# Show plot
plt.show()
```
